# Models

```python
def dnn_model():
    x_input = Input(shape=(test.shape[1],))
    x1 = Dense(units=386, activation='swish')(x_input)
    x1 = BatchNormalization()(x1)
    x1 = Dropout(rate=.45)(x1)
    
    x2 = Dense(units=192, activation='swish')(x1)
    x2 = BatchNormalization()(x2)
    x2 = Dropout(rate=.35)(x2)
    
    x3 = Dense(units=98, activation='swish')(x2)
    x3 = BatchNormalization()(x3)
    x3 = Dropout(rate=.25)(x3)
    
    x4 = Dense(units=192, activation='swish')(x3)
    x4 = BatchNormalization()(x4)
    x4 = Multiply()([x2, x4])
    x4 = Dropout(rate=.35)(x4)
    
    x5 = Dense(units=386, activation='swish')(x4)
    x5 = BatchNormalization()(x5)
    x5 = Multiply()([x1, x5])
    x5 = Dropout(rate=.45)(x5)
    
    x_e1 = Concatenate()([x3, x5])
    x = Reshape((22, 22, 1))(x_e1)
    
    x = Conv2D(16, kernel_size=(3, 3), padding='same', activation='swish')(x)
    x = AveragePooling2D((2, 2), padding='same')(x)
    x = Conv2D(8, kernel_size=(3, 3), padding='same', activation='swish')(x)
    x = AveragePooling2D((2, 2), padding='same')(x)
    x = Conv2D(8, kernel_size=(3, 3), padding='same', activation='swish')(x)
    
    enc = AveragePooling2D((2, 2), padding='same')(x)
    flat = Flatten()(enc)
    
    x = Concatenate()([x_e1, flat])
    
    x = Dense(units=128, activation='swish')(x)
    x = BatchNormalization()(x)
    x = Dropout(rate=.25)(x)
    
    x_output = Dense(units=1, activation='sigmoid')(x)
    
    model = Model(inputs=x_input, outputs=x_output, name='DNN_Model')
    
    return model
```

```python
with tf.device('/GPU:0'):
    fold = 5
    verbose = 1
    seeds = [2021, 2025]
    batch_size = 2048
    class_weight_flag = False

    counter = 0
    oof_score = 0
    y_pred_final_dnn = np.zeros((test.shape[0], 1))
    y_pred_meta_dnn = np.zeros((test.shape[0], 1))

    for sidx, seed in enumerate(seeds):
        kfold = StratifiedKFold(n_splits=fold, shuffle=True, random_state=seed)

        for idx, (train, val) in enumerate(kfold.split(X_train, y_train)):
            counter += 1
            xtrain, xtest = X_train.iloc[train, :], X_train.iloc[val, :]
            ytrain, ytest = y_train[train], y_train[val]

            zero = np.unique(ytrain, return_counts=True)[1][0]
            one = np.unique(ytrain, return_counts=True)[1][1]
            total = len(ytrain)

            weight_for_0 = (1 / zero) * (total) / 2.0
            weight_for_1 = (1 / one) * (total) / 2.0

            class_weight = {0: weight_for_0, 1: weight_for_1}
            print(f'Weight for class 0: {weight_for_0:.2f}')
            print(f'Weight for class 1: {weight_for_1:.2f}')

            model = dnn_model()
            model.compile(optimizer=Adam(learning_rate=1e-2),
                         loss='binary_crossentropy',
                         metrics=['AUC'])
            lr = ReduceLROnPlateau(monitor='val_loss', factor=.2,
                                  patience=7, verbose=verbose)
            chk_point = ModelCheckpoint(f'./Keras_DNN_Model_{counter}C.h5',
                                       monitor='val_loss', verbose=verbose,
                                       save_best_only=True, mode='min')
            es = EarlyStopping(monitor='val_loss', patience=30,
                              verbose=verbose, mode='min',
                              restore_best_weights=True)
            if not class_weight_flag:
                class_weight = None
            else:
                pass

            model.fit(xtrain, ytrain,
                     validation_data=(xtest, ytest),
                     epochs=400,
                     verbose=verbose,
                     class_weight=class_weight,
                     batch_size=batch_size,
                     callbacks=[lr, chk_point, es])
            model = load_model(f'./Keras_DMM_Model_{counter}C.h5')
            y_pred = model.predict(xtest, batch_size=batch_size)
            y_pred_meta_dnn[var] += y_pred
            y_pred_final_dnn = model.predict(X_test, bstch_size=batch_size)

            score = roc_auc_score(ytest, y_pred)
            oof_score += score
            seed_score += score

            del model, y_pred
            del xtrain, ytrain
            del xtest, ytest

            gc.collect()
```
