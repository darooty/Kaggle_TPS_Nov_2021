```python
plt.subplots(figsize=(10, 5))
g = sns.countplot(train.target)
for patch in g.patches:
    x = patch.get_x()
    height = patch.get_height()
    width = patch.get_width()
    g.text(x + width/2, height + 30,
           f'{height} / {height/train.shape[0]*100:2.2f}%',
           ha='center', weight='bold', size=13, bbox={'facecolor':'None', 'linewidth':.5}
          )
    g.spines['top'].set_visible(False)
    g.spines['right'].set_visible(False)
plt.show()
```

```python
def continuous_dist(data, var):
    plt.subplots(figsize=(20, 7))
    g = sns.distplot(data[var])
    g.set_title(f"{var}'s distribution (skew: {data[var].skew():.2f})", weight='bold', size=13)    
    plt.show()
```

```python
def continuous_dist_bytarget(data, x, y):
    f, ax = plt.subplots(1, 3, figsize=(35, 10))
    sns.violinplot(x=data[x], y=data[y], ax=ax[0], edgecolor='black', linewidth=5)
    sns.boxplot(x=data[x], y=data[y], ax=ax[1])
    sns.stripplot(x=data[x], y=data[y], ax=ax[2])
    for i in range(3):
        ax[i].spines['top'].set_visible(False)
        ax[i].spines['right'].set_visible(False)
        ax[i].set_xlabel(x, weight='bold', size=20)
        ax[i].set_ylabel(y, weight='bold', size=20)
        ax[i].set_facecolor(BACKCOLOR)
    f.suptitle(f"{y}'s distribution by {x}", weight='bold', size=25)
    plt.show()
```

```python
def search_var(data, var):
    print(data[var].describe())
    print(f'\nmissing count: {data[var].isnull().sum()}')
    continuous_dist(data, var)
    continuous_dist_bytarget(data, x='target', y=var)
```

```python
all_data = pd.concat((train, test), axis=0)
train0 =  train[train.target == 0]
train1 =  train[train.target == 1]
```

```python
b = pd.DataFrame(columns=['feature', 'data', 'type', 'count', 'mean', 'median', 'std', 'min', 'max', 'skew', 'null'])
cnt = 0
for c in train.columns:
    if c == 'id':
        continue
    b.loc[cnt] = [c, 'Train', train[c].dtype.name, train[c].count(), train[c].mean(), train[c].median(), train[c].std(), train[c].min(), train[c].max(), train[c].skew(), train[c].isnull().sum()]
    cnt+=1
    b.loc[cnt] = [c, 'All', train[c].dtype.name, all_data[c].count(), all_data[c].mean(), all_data[c].median(), all_data[c].std(), all_data[c].min(), all_data[c].max(), all_data[c].skew(), all_data[c].isnull().sum()]
    cnt+=1
    b.loc[cnt] = [c, 'Target=0', train0[c].dtype.name, train0[c].count(), train0[c].mean(), train0[c].median(), train0[c].std(), train0[c].min(), train0[c].max(), train0[c].skew(), train0[c].isnull().sum()]
    cnt+=1
    b.loc[cnt] = [c, 'Target=1', train1[c].dtype.name, train1[c].count(), train1[c].mean(), train1[c].median(), train1[c].std(), train1[c].min(), train1[c].max(), train1[c].skew(), train1[c].isnull().sum()]
    cnt+=1
b = b.set_index(['feature', 'data'],drop=True)
```
